## Build spectral dataset

### Common cores

```python
file_name = []
spe_all = []
depth_all = []
cps_all = []
core_all = []
s_depth_all = []

# open an error txt for recording errors
with open('results/error_log.txt', 'w') as ef:
    # do it core by core
    for core in xl_df.Station[~xl_df.Station.isin(pre_df.core.unique())]:
        
        # only read the 10kV which having better signal to the light elements
        spe_dir = glob.glob('data/{}/Run 1 at  10kV/*.spe'.format(core))
        
        # make sure the order follows the depthes in filename
        spe_dir.sort()
        print('{} has {} data points.'.format(core, len(spe_dir)))

        for spe in spe_dir:
                   
            check_depth = spe.split()[3].split('_')[-1]
            
            # there are some inconsistencies in nameing...as usual
            # 5 and 6 digis means in mm
            try:
                if len(check_depth) >= 5:
                    start_depth = int(check_depth)
                # 3 and 4 digit means in cm, needs to be multipled to be mm
                elif len(check_depth) >= 3:
                    start_depth = int(check_depth) * 10

                file_name.append(spe.split('/')[-1])
                
                with open(spe, 'r') as f:
                    content = []
                    lines = f.readlines()
                    for line in lines[49:]:
                        content = np.hstack((content, line.split()))
                    
                    section_depth = int(lines[13][:-3])
                    spe_all.append(content.astype(int))
                    cps_all.append(int(lines[28]))
                    core_all.append(core)
                    s_depth_all.append(section_depth)
                    depth_all.append(section_depth + start_depth)
            except ValueError :
                print('Error in filename: {}'.format(spe), file=ef)

        print('core {} is done.'.format(core))
```

No other error apart from the ValueError, which is recorded in the error_log.txt and skiped, pops up. After checking error_log.txt, I found only core PS97-083-2 has inconsistent naming spectra. The spectrum filename has TC in the position I adopt for section depth, e.g. PS97-083-2_"00TC"   20.0mm  10s  10kV 150uA No-Filter   0.0mm Run1 Rep0 Sett Low.spe. They should be identified as "Triger Core", but since there are no bulk chemistry for this TC core, I just abandon these data.

Luckily, these cores have the same spe format as the previous cores, which confirms they were scanned by the new scanning head.

### Triger cores

```python
# no initialization because I want the data to be appended
# open an error txt for appending errors
with open('results/error_log.txt', 'a') as ef:
    # do it core by core
    # these are the cores having TC as a subfolder
    for core in ['PS97-084-1', 'PS97-085-3', 'PS97-089-1', 'PS75-093-1', 'PS75-083-1']:
        
        # only read the 10kV which having better signal to the light elements
        spe_dir = glob.glob('data/{}/TC/Run 1 at  10kV/*.spe'.format(core))
        
        # make sure the order follows the depthes in filename
        spe_dir.sort()
        print('{} has {} data points.'.format(core, len(spe_dir)))

        for spe in spe_dir:
                   
            check_depth = spe.split()[3].split('_')[-1]
            
            # there are some inconsistencies in nameing...as usual
            # 5 and 6 digis means in mm
            try:
                if len(check_depth) >= 5:
                    start_depth = int(check_depth)
                # 3 and 4 digit means in cm, needs to be multipled to be mm
                elif len(check_depth) >= 3:
                    start_depth = int(check_depth) * 10
            
            except ValueError:
                # There is only PS75-093-1 has TC in the naming and their sectiondepth should
                # all be starting at 0
                start_depth = 0

            file_name.append(spe.split('/')[-1])
            
            with open(spe, 'r') as f:
                content = []
                lines = f.readlines()
                for line in lines[49:]:
                    content = np.hstack((content, line.split()))
                
                section_depth = int(lines[13][:-3])
                spe_all.append(content.astype(int))
                cps_all.append(int(lines[28]))
                core_all.append(core+'_TC')   # mark as the triger cores
                s_depth_all.append(section_depth)
                depth_all.append(section_depth + start_depth)
            
        print('core {} is done.'.format(core))
```

```python
spe_df = pd.DataFrame(spe_all, columns = [str(_) for _ in range(2048)])
spe_df['cps'] = cps_all
spe_df['core'] = core_all
spe_df['composite_depth_mm'] = depth_all
spe_df['section_depth_mm'] = s_depth_all
spe_df['filename'] = file_name

print(spe_df.head())
print(spe_df.shape)
```

```python
mask_c = spe_df.columns[:2048]  # only the channels
merge_df = pd.DataFrame()

for _, row in bulk_df.iterrows():
    mid = row['mid_depth_mm']
    core = row['core']
    
    # get the spe in 10 mm interval
    mask_r = ((spe_df.composite_depth_mm >= (mid-5)) & 
              (spe_df.composite_depth_mm <= (mid+5)) & 
              (spe_df.core == core))
    merge_df = pd.concat(
        [merge_df, spe_df.loc[mask_r, mask_c].apply(np.mean, axis = 0).append(row)],
        axis = 1
    )
    
merge_df = merge_df.T.reset_index(drop = True)
```
