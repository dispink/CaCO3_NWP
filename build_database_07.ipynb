## Build spectral dataset
I've manually check the scanning info of these cores to the rest cores. They are relevant.

### SO264-69-2

```python
file_name = []
spe_all = []
depth_all = []
cps_all = []
core_all = []
s_depth_all = []

core = 'SO264-69-2'
# only read the 10kV which having better signal to the light elements
# BTW, its naming in the spe files is miswritten in SO264-69-1
spe_dir = glob.glob('data/SO264-69-2/Run 1 at  10kV/*.spe')

# make sure the order follows the depthes in filename
spe_dir.sort()
print('{} has {} data points.'.format(core, len(spe_dir)))

for spe in spe_dir:
    check_depth = spe.split()[3].split('_')[-1]

    # 5 and 6 digis means in mm
    if len(check_depth) >= 5:
        start_depth = int(check_depth)
    # 3 and 4 digit means in cm, needs to be multipled to be mm
    elif len(check_depth) >= 3:
        start_depth = int(check_depth) * 10

    file_name.append(spe.split('/')[-1])

    with open(spe, 'r') as f:
        content = []
        lines = f.readlines()
        for line in lines[49:]:
            content = np.hstack((content, line.split()))
        
        section_depth = int(lines[13][:-3])
        spe_all.append(content.astype(int))
        cps_all.append(int(lines[28]))
        core_all.append(core)
        s_depth_all.append(section_depth)
        depth_all.append(section_depth + start_depth)

print('core {} is done.'.format(core))
```

```python
# I rename the core name to avoid confusion with the data of previous version
core = 'LV28-44-3-n'

# only read the 10kV which having better signal to the light elements
spe_dir = glob.glob('data/LV28-44-3(554-1113cm)/Run 1 at  10kV_554-end/*.spe')

# make sure the order follows the depthes in filename
spe_dir.sort()
print('{} has {} data points.'.format(core, len(spe_dir)))

for spe in spe_dir:
    
    check_depth = spe.split()[3].split('_')[-1]

    # 5 and 6 digis means in mm
    if len(check_depth) >= 5:
        start_depth = int(check_depth)
    # 3 and 4 digit means in cm, needs to be multipled to be mm
    elif len(check_depth) >= 3:
        start_depth = int(check_depth) * 10

    file_name.append(spe.split('/')[-1])

    with open(spe, 'r') as f:
        content = []
        lines = f.readlines()
        for line in lines[49:]:
            content = np.hstack((content, line.split()))
        
        section_depth = int(lines[13][:-3])
        spe_all.append(content.astype(int))
        cps_all.append(int(lines[28]))
        core_all.append(core)
        s_depth_all.append(section_depth)
        depth_all.append(section_depth + start_depth)

print('core {} is done.'.format(core))
```

```python
spe_df = pd.DataFrame(spe_all, columns = [str(_) for _ in range(2048)])
spe_df['cps'] = cps_all
spe_df['core'] = core_all
spe_df['composite_depth_mm'] = depth_all
spe_df['section_depth_mm'] = s_depth_all
spe_df['filename'] = file_name

spe_df.info()
```

```python
spe_df[spe_df.isna().any(axis=1)]
```

### Build composite_id

```python
composite_id = []
for core, depth in zip(spe_df.core, spe_df.composite_depth_mm):
    composite_id.append('{}_{:05}'.format(core, depth))
    
spe_df['composite_id'] = composite_id
```

### Drop duplicates

```python
clean_df = spe_df.drop_duplicates('composite_id', keep = 'last')
len(clean_df)
```

#### Duplicates due to the folder naming
This issue was solved, but I leave a note here. I found it because almost all the data in LV28-44-3 are duplicated due to the duplicated composite depth. After checking, I realized there is a space in the folder name, which causes the split() function catching wrong start_depth.

Now the duplicates are reasonable, which means they are just the duplucating measurements at the begining and end of the core sections.

```python
spe_df.loc[spe_df.composite_id.duplicated(), ['core', 'composite_depth_mm','section_depth_mm', 'filename']].to_csv('results/check_du_{}.csv'.format(date))
```

### Build section no.

```python
clean_df = clean_df.set_index('composite_id')

section_all = []
# make sure the order follows the core and composite depth
clean_df.sort_values(by = 'composite_id', axis = 0, inplace = True)

for core in np.unique(clean_df.core):    
    # I assume every core scanned from section 0 so the first section in the core 
    # is marked as section 0. The deeper the larger number
    section = 0
    X = clean_df.loc[clean_df.core == core, 'section_depth_mm']
    for i in range(len(X)):
        section_all.append(section)
        try:
            # when section changes, the section depth should be reset to smaller number
            if X[i] > X[i + 1]:
                section += 1
        except IndexError:
            print('bottom of the core {}'.format(core))
        
clean_df['section'] = section_all
```

## Build merged spe and bulk dataset

```python
mask_c = spe_df.columns[:2048]  # only the channels
merge_df = pd.DataFrame()

for _, row in bulk_df.iterrows():
    mid = row['mid_depth_mm']
    core = row['core']
    
    # get the spe in 10 mm interval
    mask_r = ((spe_df.composite_depth_mm >= (mid-5)) & 
              (spe_df.composite_depth_mm <= (mid+5)) & 
              (spe_df.core == core))
    # make sure there are spe data in the depth range
    # otherwise, it gives you NA
    if mask_r.sum() > 0:
        merge_df = pd.concat(
            [merge_df, spe_df.loc[mask_r, mask_c].apply(np.mean, axis = 0).append(row)],
            axis = 1
        )
    
merge_df = merge_df.T.reset_index(drop = True)
```

# Build Ca & Br datasets
